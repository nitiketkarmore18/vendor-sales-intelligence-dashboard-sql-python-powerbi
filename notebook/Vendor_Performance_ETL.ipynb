```python
import pandas as pd
import numpy as np
```

# Data Extraction

```python
sales = pd.read_csv('/content/sales.csv')

purchases = pd.read_csv('/content/purchases.csv')

begin_inventory = pd.read_csv('/content/begin_inventory.csv')

end_inventory = pd.read_csv('/content/end_inventory.csv')

purchase_prices = pd.read_csv('/content/purchase_prices.csv')

vendor_invoice = pd.read_csv('/content/vendor_invoice.csv')

print("All datasets extracted successfully ✅")
```

# Dataset Validation

```python
datasets = {
    "sales": sales,
    "purchases": purchases,
    "begin_inventory": begin_inventory,
    "end_inventory": end_inventory,
    "purchase_prices": purchase_prices,
    "vendor_invoice": vendor_invoice
}

for name, df in datasets.items():

    print("\n", "="*50)
    print(f"Dataset: {name.upper()}")
    print("="*50)

    print("Rows:", df.shape[0])
    print("Columns:", df.shape[1])
```

## Validation Observation

The extracted datasets contain large-scale operational business records suitable for enterprise-level ETL workflows, KPI engineering, dashboard analytics, and business intelligence reporting.

# Missing Value Validation

```python
for name, df in datasets.items():

    print("\n", "="*50)
    print(f"Missing Values in {name.upper()}")
    print("="*50)

    print(df.isnull().sum())
```

## ETL Observation

The datasets contain minimal missing values. Missing records mainly appear in vendor approval and location-related fields, representing minor operational inconsistencies in business transactions.

# Duplicate Record Validation

```python
for name, df in datasets.items():

    print("\n", "="*50)
    print(f"Duplicate Records in {name.upper()}")
    print("="*50)

    print("Duplicate Rows:", df.duplicated().sum())
```

## ETL Observation

Duplicate validation ensures transactional consistency and prevents inflated KPI calculations, duplicate revenue generation, and inventory inaccuracies during analytical processing.

# Datatype Standardization

```python
sales['SalesDate'] = pd.to_datetime(
    sales['SalesDate']
)

purchases['PODate'] = pd.to_datetime(
    purchases['PODate']
)

vendor_invoice['InvoiceDate'] = pd.to_datetime(
    vendor_invoice['InvoiceDate']
)

print("Date columns standardized successfully ✅")
```

```python
sales.info()

purchases.info()

vendor_invoice.info()
```

## Transformation Observation

Date standardization enables time-series analysis, monthly KPI generation, forecasting preparation, and dashboard compatibility for business intelligence workflows.

# Missing Value Handling

```python
sales = sales.dropna().copy()

purchases = purchases.dropna().copy()

begin_inventory = begin_inventory.dropna().copy()

end_inventory = end_inventory.dropna().copy()
```

```python
print("Sales Null Values:",
      sales.isnull().sum().sum())

print("Purchases Null Values:",
      purchases.isnull().sum().sum())

print("Begin Inventory Null Values:",
      begin_inventory.isnull().sum().sum())

print("End Inventory Null Values:",
      end_inventory.isnull().sum().sum())
```

## Cleaning Observation

The datasets were permanently cleaned to remove incomplete operational records and improve analytical consistency for downstream dashboarding and KPI workflows.

# Feature Engineering

```python
sales['Revenue_Per_Unit'] = (
    sales['SalesDollars']
    / sales['SalesQuantity']
)

sales['Year'] = sales['SalesDate'].dt.year

sales['Month'] = sales['SalesDate'].dt.month

sales['Month_Name'] = sales['SalesDate'].dt.month_name()

sales['Day'] = sales['SalesDate'].dt.day
```

```python
begin_inventory['InventoryValue'] = (
    begin_inventory['onHand']
    * begin_inventory['Price']
)

end_inventory['InventoryValue'] = (
    end_inventory['onHand']
    * end_inventory['Price']
)
```

## Feature Engineering Observation

Additional KPI and analytical fields were generated to support inventory valuation, time-series analysis, profitability analytics, and dashboard-ready business intelligence reporting.

# Aggregation Pipelines



## Vendor Revenue Summary Table

```python
vendor_summary = sales.groupby(
    'VendorName'
).agg({

    'SalesDollars':'sum',
    'SalesQuantity':'sum',
    'Revenue_Per_Unit':'mean'

}).reset_index()
```

```python
vendor_summary.rename(columns={

    'SalesDollars':'Total_Revenue',
    'SalesQuantity':'Total_Quantity_Sold',
    'Revenue_Per_Unit':'Average_Revenue_Per_Unit'

}, inplace=True)
```

```python
vendor_summary.head()
```

## Aggregation Observation

The vendor summary table provides vendor-level KPI metrics suitable for dashboard reporting, vendor ranking, profitability analysis, and procurement intelligence workflows.

## Product Performance Summary Table

```python
product_summary = sales.groupby(
    'Description'
).agg({

    'SalesDollars':'sum',
    'SalesQuantity':'sum',
    'Revenue_Per_Unit':'mean'

}).reset_index()
```

```python
product_summary.rename(columns={

    'SalesDollars':'Total_Product_Revenue',
    'SalesQuantity':'Total_Product_Quantity',
    'Revenue_Per_Unit':'Average_Revenue_Per_Unit'

}, inplace=True)
```

```python
product_summary.head()
```

## Store Performance Summary Table

```python
store_summary = sales.groupby(
    'Store'
).agg({

    'SalesDollars':'sum',
    'SalesQuantity':'sum'

}).reset_index()
```

```python
store_summary.rename(columns={

    'SalesDollars':'Total_Store_Revenue',
    'SalesQuantity':'Total_Store_Quantity'

}, inplace=True)
```

```python
store_summary.head()
```

## Inventory Summary Table

```python
inventory_summary = end_inventory.groupby(
    'Description'
).agg({

    'onHand':'sum',
    'InventoryValue':'sum'

}).reset_index()
```

```python
inventory_summary.rename(columns={

    'onHand':'Remaining_Stock',
    'InventoryValue':'Remaining_Inventory_Value'

}, inplace=True)
```

```python
inventory_summary.head()
```

## ETL Aggregation Observation

Aggregated analytical tables significantly reduce dashboard processing complexity and improve Power BI performance by transforming raw transactional records into summarized KPI-ready datasets.
